```python
def getaddresscode(x):
    try:
        result = requests.get(url_juso.format(x)).json()["results"]["juso"][0]["zipNo"]
        print(result)
        return result
    except:
        print("error occurred")
        return 0





data["new_post"] = data["주소"].apply(getaddresscode)
        


```
