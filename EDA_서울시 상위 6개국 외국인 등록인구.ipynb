```python
import pandas as pd
file_dir = "C://Users/Youngju Hong/Documents/projectfile/"
```

```python
##1. 주민등록인구 핸들링 (외국인 주민인구의 경우 (국적이 병기되어 있는 경우) 구단위 인구밖에 존재하지 않음)
```

```python
local_data = pd.read_table(f"{file_dir}서울시 주민등록인구 (구별) 통계.txt", sep = '\t')
local_data = local_data.iloc[3:len(local_data), [1,3]]
local_data.columns = ["gu", "population"]
local_data.head()
```

```python
#생활인구
foreign_data = pd.read_excel(f"{file_dir}foreign_living_population_by_gu.xlsx")
foreign_data.head()
```

```python
#생활인구
merged = pd.merge(local_data, foreign_data, on = "gu")
del merged["Unnamed: 0"]
merged.head()
```

```python
#주민등록인구
foreign_living_data = pd.read_table(f"{file_dir}foreign_reg_population.txt")
foreign_living_data = foreign_living_data.replace({',': '', ' ': '', '-': 0}, regex = True)
foreign_living_data.columns = foreign_living_data.columns.str.replace(' ', '')
foreign_living_data = foreign_living_data[foreign_living_data["성별"] == "계"]
```

```python
#foreign_living_data (주민등록인구) 데이터 처리
foreign_living_data["chn_reg"] = foreign_living_data.loc[:, ["한국계중국인","중국(한국계중국인제외)"]].astype(float).sum(axis = 1) #형변환
foreign_data_revised = foreign_living_data[["자치구", "합계", "미국", "타이완", "일본", "베트남", "캐나다", "chn_reg"]] #특정 열 선택
foreign_data_revised
```

```python
foreign_data_revised = foreign_data_revised.rename({"미국": "usa_reg", "타이완": "twn_reg", "일본": "jpn_reg", "캐나다": "can_reg", "베트남": "vnm_reg"}, axis = 1)
foreign_data_revised["major_sum"] = foreign_data_revised.loc[:,"usa_reg":"chn_reg"].astype(float).sum(axis = 1)
foreign_data_revised["etcsum"] = foreign_data_revised["합계"].astype(float).subtract(foreign_data_revised["major_sum"])
foreign_data_revised = foreign_data_revised.rename(columns = {"자치구": "gu", "etcsum": "etcsum_reg"})
foreign_data_revised
```

```python
merged = pd.merge(merged, foreign_data_revised, on = "gu")
merged.to_excel("final.xlsx")
```
