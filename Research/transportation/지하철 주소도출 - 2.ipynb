```python
import pandas as pd
import requests
file_dir_mac = "/Users/hong/Documents/GitHub/seoulresearch_project/Research/"
```

```python
address_file = pd.read_excel(f"{file_dir_mac}/Addresses.xlsx")
```

```python
address_file.head()
```

```python
address_file.columns = ["Results", "Address"]
address_file.head()
```

```python
def split_with_parentheses(name):
    return name[name.find("(")+1: len(name)-1]

def retrieve_simplified_address(address):
    if "nan" in str(address):
        return
    else:
        print(address)
        temp = str(address).split(' ')
        print(temp)

        return (lambda x: " ".join(x[:3]) if len(x) >= 3 else " ".join(x))(temp)


address_file["station"] = address_file["Results"].apply(split_with_parentheses)
address_file["line"] = address_file["Results"].apply(lambda x: str(x).split(' ')[0])
address_file["city"] = address_file["Address"].apply(lambda x: str(x).split(' ')[0])
address_file["address_simplified"] = address_file["Address"].apply(retrieve_simplified_address)
```
