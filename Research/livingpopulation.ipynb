```python
import pandas as pd
foreign_data = pd.read_table("C:\\Users\Youngju Hong\Documents\projectfile\SE_SPOP_FORN_LONG_RESD_DONG_2020년1월-5월.txt", sep = ",")
foreign_data["ymd"] = pd.to_datetime(foreign_data["ymd"], format = "%Y%m%d")
foreign_data.head()
```

```python
foreign_data_grouped = foreign_data.groupby(by = ["tt", "h_dng_cd"]).mean().reset_index() #group by time and dong code
foreign_data_grouped["gu_code"] = foreign_data_grouped["h_dng_cd"].apply(lambda x: str(x)[:4]) #extract gu code
foreign_data_grouped
```

```python
district_code = pd.read_csv("C:\\Users\Youngju Hong\Documents\projectfile\행정동코드.csv")
district_code.columns = ["통계청코드", "행자부코드", "시도명", "구", "동"]
district_code["구코드"] = district_code["행자부코드"].apply(lambda x: str(x)[:4])
district_code["구코드"]
```

```python
district_code_gu = dict(zip(district_code["구코드"], district_code["구"]))
district_code_dong = dict(zip(district_code["행자부코드"], district_code["동"]))
hangja_to_tong = dict(zip(district_code["행자부코드"], district_code["통계청코드"]))

#print(district_code_gu)
print(foreign_data_grouped["h_dng_cd"])
```

```python
foreign_data_grouped["gu"] = foreign_data_grouped["gu_code"].apply(lambda x: district_code_gu[x])
foreign_data_grouped["dong"] = foreign_data_grouped["h_dng_cd"].apply(lambda x: district_code_dong[x])
foreign_data_grouped["h_dng_cd_tong"] = foreign_data_grouped["h_dng_cd"].apply(lambda x: hangja_to_tong[x])
foreign_data_grouped["time"] = foreign_data_grouped["tt"].apply(lambda x: "Day" if (x>=9 and x<=20) else "Night")
foreign_data_grouped.head()
```

```python
#stat_temp.to_excel("foreign_living_population_by_dong.xlsx")
stat_temp = foreign_data_grouped.groupby(["h_dng_cd_tong","time"])[["fornlong_spop","chn","vnm","usa","jpn","can","twn","etcsum"]].mean().reset_index()
print(stat_temp)
stat_temp_day = stat_temp[stat_temp["time"] == "Day"]
stat_temp_night = stat_temp[stat_temp["time"] == "Night"]

#stat_temp = stat_temp.groupby(by = ["gu", "time"]).sum().reset_index()
#stat_temp.to_excel("foreign_living_population_by_gu.xlsx")
```

```python
#For GIS
gis = pd.read_excel("C:\\Users\Youngju Hong\Documents\projectfile\gis.xlsx", sheet_name = "GIS version")
gis = gis.rename(columns = {"ADM_CD": "h_dng_cd_tong"})
day_population = pd.merge(gis, stat_temp_day, on = "h_dng_cd_tong", how = "left")
night_population = pd.merge(gis, stat_temp_night, on = "h_dng_cd_tong", how = "left")
```

```python
##2. Local Population
#instate handling

local_data = pd.DataFrame(columns = ["기준일ID", "시간대구분", "행정동코드", "총생활인구수"])
#"기준일ID", "시간대구분", "행정동코드", "총생활인구수"


for i in range(1,6):
    pop = pd.read_csv(f"C:\\Users\Youngju Hong\Documents\projectfile\동별내국인생활인구_2020년_1월_5월\LOCAL_PEOPLE_DONG_20200{str(i)}.csv")
    local_data = pd.concat([local_data, pop])
    #grouped = pop.groupby(by = ["행정동코드"], as_index = False)
    #grouped
```

```python
local_data["time"] = local_data["시간대구분"].apply(lambda x: "Day" if (x>=9 and x<=20) else "Night") #day (09-20) and night(20-08)
#local_data["dong"] = local_data["행정동코드"].apply(lambda x: district_code_dong[x])
local_data["gu"] = local_data["행정동코드"].apply(lambda x: district_code_gu[str(x)[:4]])
local_data["h_dng_cd_tong"] = local_data["행정동코드"].apply(lambda x: hangja_to_tong[x])

stat_temp = local_data.groupby(["h_dng_cd_tong", "time"])[["총생활인구수"]].mean().reset_index()
day_population = pd.merge(day_population, stat_temp[stat_temp["time"] == "Day"], on = "h_dng_cd_tong", how = "left")
night_population = pd.merge(night_population, stat_temp[stat_temp["time"] == "Night"], on = "h_dng_cd_tong", how = "left")
```
