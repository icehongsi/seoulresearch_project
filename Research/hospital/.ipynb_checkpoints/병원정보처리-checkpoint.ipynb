```python
#Option 1. 좌표변환
from pyproj import Proj, Transformer, CRS
from functools import partial
import pandas as pd
import requests
import time
def transformation(x_coordinate, y_coordinate, epsg_from = 4326, epsg_to = 5179):
    return Transformer.from_crs(f"epsg:{epsg_from}", f"epsg:{epsg_to}").transform(y_coordinate, x_coordinate)

def transformation_df(x_df, y_df, epsg_from = 4326, epsg_to = 5179):
    for i in range(len(x_df)):
        if x_df[i] == np.nan:
            x_df[i], y_df[i] = None, None
        else:         
            x_df[i], y_df[i] = Transformer.from_crs(f"epsg:{epsg_from}", f"epsg:{epsg_to}").transform(y_df[i], x_df[i])
            print(x_df[i], y_df[i])
    return pd.DataFrame(data = {"X_transformed": x_df, "Y_transformed": y_df})
```

```python
data2 = transformation_df(hospital["좌표정보(X)"], hospital["좌표정보(Y)"], epsg_from = 2097)
hospital["좌표정보(X)"], hospital["좌표정보(Y)"] = data2["X_transformed"], data2["Y_transformed"]
```
