```python
jiga_insight= jiga.groupby(by = ["법정동코드", "법정동명", "공동주택구분명"])["평방미터당 가격"].agg(["median", "mean", "max", "min", "std", "skew"]).reset_index()

```

```python
trimmean = lambda x: scipy.stats.trim_mean(x, 0.1)

jiga_insight= jiga.groupby(by = ["법정동코드", "법정동명", "공동주택구분명"])["평방미터당 가격"].agg(["median", "mean", "max", "min", "std", "skew"]).reset_index()


jiga_insight
```

```python
jiga_insight["10% trimmed mean"] = jiga.groupby(by = ["법정동코드", "법정동명", "공동주택구분명"])["평방미터당 가격"].agg(trimmean).reset_index()["평방미터당 가격"]
```
