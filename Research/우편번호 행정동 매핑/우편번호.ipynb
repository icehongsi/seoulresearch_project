```python
import pandas as pd
```

```python
data = pd.read_excel("C://Users/Youngju Hong/Documents/GitHub/seoulresearch_project/Research/Untitled Folder/grouped.xlsx")
data = data.reset_index()
print(data)
```

```python
data.sort_values(by = ["관리번호"], ascending = False)
```

```python
#조건 1. 같은 동 + 같은 우편번호
for i in range(len(data)):
    if data.iloc[i,3] == 0:
        post, dong = data.iloc[i,4], data.iloc[i,5]
        try:
            df = data[(data.iloc[:, 4] == post) & (data.iloc[:, 5] == dong)].sort_values(by = ["관리번호"], ascending = False)
            data.iloc[i,3] = df.iloc[0,3]
            #data.iloc[i,2] = df["행정동명"]
        except:
            pass
    
```
