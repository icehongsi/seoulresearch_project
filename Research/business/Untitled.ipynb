```python
data = pd.read_table("서울시 금융기관 (동별) 통계.txt", header = 2, )
```

```python
data = data.replace("-", "0")
```

```python
data.loc[:, "한국은행":"외국은행"] = data.loc[:, "한국은행":"외국은행"].astype(int)
```

```python
data = data[(data["동"] != "소계") & (data["동"] != "합계")]
data.columns
```

```python
data["구동"] = data["자치구"] + " " + data["동"]
```

```python
data
```

```python
data_aggregated = data.groupby(by = "구동").filter(lambda x: True).reset_index()
data_aggregated
data_aggregated["구동"] = data_aggregated["구동"].apply(lambda x: x.replace("·", ".") if "·" in x else x)
```
