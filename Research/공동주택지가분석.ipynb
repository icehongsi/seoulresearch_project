```python
dict1 = dict(zip(code_info["주소"], code_info["관리번호"]))
dict2 = dict(zip(code_info["관리번호"], code_info["행정동명"]))
dict3 = {}

for i in dict1.keys():
    dict3[i] = dict2[dict1[i]]

# 주소 - 행정동명간 맵핑
hangdong_to_code = dict(zip(code_info["행정동명"], code_info["행정동코드"]))
```
