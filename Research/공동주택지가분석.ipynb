```python
jiga = jiga[jiga["기준년도"] == "2020"] # Retrieve Information only for the one from 2020
```

```python
# subset data
jiga["공시가격"] = jiga["공시가격"].astype(float) #공시가격은 float type으로
#jiga["전용면적"] = jiga["전용면적"].astype(float)
jiga["특수지구분명"] = jiga["특수지구분명"].apply(lambda x: "" if x == "일반" else "산 ") #일반용지일 경우 결측값
```

```python
code_info = pd.read_excel("C:\\Users\Youngju Hong\Documents\projectfile\지역코드\부동산종합데이터.xlsx", sheet_name = 0, dtype = object)
```

```python
dict1 = dict(zip(code_info["주소"], code_info["관리번호"]))
dict2 = dict(zip(code_info["관리번호"], code_info["행정동명"]))
dict3 = {}

for i in dict1.keys():
    dict3[i] = dict2[dict1[i]]

# 주소 - 행정동명간 맵핑
hangdong_to_code = dict(zip(code_info["행정동명"], code_info["행정동코드"]))
```

```python
hangdong_to_code
```

```python
def find_hangdong(x):
    try:
        return dict3[x]
    except:
        return None
    
jiga["행정동명"] = jiga["주소"].apply(find_hangdong)
```

```python
jiga = jiga[~jiga["행정동명"].isna()] #결측값이 18개이므로 단순하게 결측값을 버립니다
```

```python
jiga["평방미터당 가격"] = jiga["공시가격"] / jiga["전용면적"].astype(float)
```

```python
jiga["평방미터당 가격"]
```

```python
jiga_insight= jiga.groupby(by = ["행정동명", "공동주택구분명"])["평방미터당 가격"].agg(["median", "mean", "max", "min", "std", "skew"]).reset_index()
jiga_insight
```

```python
import scipy.stats
trimmean = lambda x: scipy.stats.trim_mean(x, 0.1)
jiga_insight["10% trimmed mean"] = jiga.groupby(by = ["행정동명", "공동주택구분명"])["평방미터당 가격"].agg(trimmean).reset_index()["평방미터당 가격"]
```

```python
def hangdong_to_codes(x):
    return hangdong_to_code[x]

jiga_insight["행정동코드"] = jiga_insight["행정동명"].apply(hangdong_to_codes).astype(str)
```
